```python
import geopy.distance

# Coordonnée à tester
coordonnees_borne = [1.407644, 43.628261]  # Coordonnée de la borne à tester (lat, lon)

# Coordonnée réelle (par exemple, à partir de Google Maps ou d'une autre source fiable)
coordonnees_reelle = [1.4060, 43.6290]  # Exemple d'une coordonnée "réelle"

# Calcul de la distance entre les deux coordonnées
distance = geopy.distance.distance(coordonnees_borne, coordonnees_reelle).km

print(f"La distance entre la borne et la position réelle est de {distance:.4f} km.")
```

```python
from geopy.geocoders import Nominatim

# Initialiser le géocodeur avec un user_agent personnalisé
geolocator = Nominatim(user_agent="mon_application_unique_1234")

# Coordonnée à tester (latitude, longitude)
coordonnees = (43.628261, 1.407644)

try:
    # Faire la géocodification inversée
    location = geolocator.reverse(coordonnees, timeout=10)

    # Afficher l'adresse si elle existe
    if location:
        print(location.address)
    else:
        print("Aucune adresse trouvée pour ces coordonnées.")
except Exception as e:
    print(f"Une erreur est survenue : {e}")
```

```python
import pandas as pd

# Chargement du fichier CSV
df = pd.read_csv('consolidation-etalab-schema-irve-statique-v-2.3.1-20250321.csv')
```

```python
df.head()
```

```python
# Vérifier les valeurs uniques dans la colonne 'accessibilite_pmr'
df['accessibilite_pmr'].unique()
```

```python
# Trouver les lignes avec des valeurs manquantes dans la colonne 'accessibilite_pmr'
df_missing_pmr = df[df['accessibilite_pmr'].isna()]

# Afficher les lignes avec des valeurs manquantes
df_missing_pmr
```

```python
# Filtrer les lignes où 'accessibilite_pmr' ne contient pas "Oui", "Non", ou "Inaccessible"
df_incoherent_pmr = df[~df['accessibilite_pmr'].isin(['Oui', 'Non', 'Inaccessible'])]

# Afficher ces lignes pour analyse
df_incoherent_pmr
```

```python
# Compter le nombre de stations réservées aux PMR (accessibles)
nb_stations_pmr = df[df['accessibilite_pmr'] == 'Oui'].shape[0]

# Afficher le nombre de stations accessibles PMR
print(f"Nombre de stations accessibles PMR : {nb_stations_pmr}")
```

```python
# Compter le nombre de stations accessibles aux PMR (accessibles)
nb_stations_accessibles_pmr = df[df['accessibilite_pmr'] == 'Oui'].shape[0]

# Afficher le nombre de stations accessibles aux PMR
print(f"Nombre de stations accessibles aux PMR : {nb_stations_accessibles_pmr}")
```
